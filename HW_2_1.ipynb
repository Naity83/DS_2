*   Вивести перші рядки таблиці за допомогою методу head




```python
import pandas as pd
import numpy as np
```

```python
table = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands=".")
df = table[0]
df.head()
```

*   Визначте кількість рядків та стовпців у датафреймі (атрибут shape)




```python
df.shape
```



*   Замініть у таблиці значення "—" на значення NaN




```python
replace = df.replace('—', np.NaN)
replace
```



*   Визначте типи всіх стовпців за допомогою dataframe.dtypes




```python
df.dtypes
```



*  Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"




```python
df[df.columns[1:]] = df[df.columns[1:]].apply(pd.to_numeric, errors='coerce')
df.dtypes
```



*   Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)




```python
df.isnull().sum()
```



*   Видаліть з таблиці дані по всій країні, останній рядок таблиці




```python
df.drop(len(df)-1)
```



*   Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)




```python
df_mean=df.fillna(df.mean(numeric_only=True))
df_mean
```



*   Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні




```python
ukraine = df['2019'][27]
result = df[df['2019'] > ukraine]
result
```



*   У якому регіоні була найвища народжуваність у 2014 році?




```python
index = df['2014'].idxmax()
region = df.loc[index, 'Регіон']
region
```



*   Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році




```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.bar(df['Регіон'].loc[0:26], df['2019'].loc[0:26], color=['pink'])
plt.title('Діаграма народжуваності по регіонах у 2019 році')
plt.xlabel('Регіони')
plt.ylabel('Коефіціент народжуваності')
plt.xticks(rotation=90)
plt.legend(['2019 рік'])
plt.show()
```

```python
kiiv = df[df['Регіон'] == 'Київська']
plt.plot(kiiv.columns[1:], kiiv.iloc[0,1:], marker='o', markerfacecolor='darkblue')
plt.title('Динаміка народжуваності у Київській області')
plt.xlabel('Рік')
plt.ylabel('Кількість народжених (на 1000 осіб)')
plt.show()
```

```python
plt.pie(
    df_mean['2019'].loc[0:26],
    labels=df_mean['Регіон'].loc[0:26],
    shadow=True,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)
plt.title('Динаміка народжуваності по областям за 2019 рік', pad=40)
plt.show()
```
