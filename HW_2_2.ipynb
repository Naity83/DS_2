

*   Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"




```python
data = df
for column in data.columns:
  if data[column].isnull().sum() > 0 and column != 'Мова програмування':
    print(column)
    data = data.drop(column, axis=1)
data.head()
```



*   Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python




```python
python_data = data[data["Мова програмування"] == "Python"]
python_data
```



*   Визначте розмір таблиці python_data за допомогою методу shape



*   Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"




```python
gr_by_posada = python_data.groupby(by = 'Посада')
gr_by_posada
```



*   Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"




```python
salary = gr_by_posada['Зарплата на місяць'].agg(['min', 'max'])
salary
```



*   Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"




```python
def fill_avg_salary(salary):
  return salary.mean()

salary['avg'] = salary.apply(fill_avg_salary, axis=1)
salary
```



*   Створіть описову статистику за допомогою методу describe для нового стовпчика.




```python
salary['avg'].describe()
```



*   Збережіть отриману таблицю в CSV файл




```python
salary.to_csv('salary_new_data.csv')
```

```python
plt.pie(
    salary['min'],
    labels=salary.index,
    shadow=True,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)
plt.title('Мінімальна заробітна плата для мови Python', pad=40)
plt.show()
```

```python

position = list(salary.index)

salaries_1 = list(salary['min'].sort_values(ascending=True))

salaries_2 = list(salary['max'].sort_values(ascending=True))

df_plot = pd.DataFrame({ 'min': salaries_1, 'max': salaries_2}, index=position)
df_plot.plot.barh(color=['lightsteelblue', 'g', 'orange'])
plt.title('Salary')
plt.xlabel('Зарплата, $', fontsize='small', color='midnightblue')
plt.ylabel('Посада', fontsize='small', color='midnightblue')
plt.show()
```
